```python
import csv
import glob
import shutil
import os
import pandas as pd
import time, os, fnmatch, shutil
import openpyxl
import numpy as np

# Read .csv from Downloads folder and get the reporting month
download_path = 'C://Users//nnagaraj//Downloads/t_worc_families.csv'
os.chdir('C://Users//nnagaraj//Downloads/')
f = pd.read_csv(download_path)


#Create variable timestamp 
t = time.localtime()
timestamp = time.strftime('%Y%m%d_%H%M', t)


# Create folder with format (201907) in H: drive
reporting_month = str(f.loc[0,'reporting_month'])
folder_name = reporting_month

dir_name = 'H://Current_Working_Folder//Reports Analytics//'
os.chdir(dir_name)

if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    
path = 'C://Users//nnagaraj//Downloads'
os.chdir(path)
source = os.listdir(path)
destination = os.path.join(dir_name, folder_name)

# move t_worc_children.csv from Downloads to destination
for files in source:
    if files.endswith("t_worc_families.csv"):
        shutil.move(files, destination)
        
os.chdir(destination)      

# Read csv
dff = pd.read_csv('t_worc_families.csv')
dff = pd.DataFrame(dff)


dff = dff.rename({'tem21a_amt':'cash & cash equivalents_amount',
                 'item21b_nbr_mth':'cash & cash equivalents_nbr_month',
                 'item22a_amt':'tanf child care_amount',
                 'item22b_child_cover':'tanf child care_children_covered',
                 'item22c_nbr_mth':'tanf child care_nbr_months',
                 'item23a_amt':'transportation_amount',
                 'item23b_nbr_mth':'transportation_nbr_months',
                 'item24a_amt':'transitional services_amount',
                 'item24b_nbr_mth':'transitional services_nbr_months',
                 'item25a_amt':'other_amount',
                 'item25b_nbr_mth':'other_nbr_months',
                 'item26a_sanc_red':'reason for amount of reduction_sanction reduction_amount',
                 'item26a_work_req':'reason for amount of reduction_work requirements sanction',
                 'item26a_no_hsd':'reason for amount of reduction_Family Sanction for an Adult with No High School Diploma or Equivalent',
                 'item26a_teen_parent':'reason for amount of reduction_Sanction for Teen Parent not Attending School',
                 'item26a_chid_sup':'reason for amount of reduction_Non-Cooperation with Child Support',
                 'item26a_resp_plan':'reason for amount of reduction_Failure to Comply with an Individual Responsibility Plan',
                 'item26a_other':'reason for amount of reduction_Other Sanction',
                 'item26b_recoup':'reason for amount of reduction_Recoupment of Prior Overpayment',
                 'item26c_other_red':'other_other total reduction amount:',
                 'item26c_family_cap':'other_Family Cap',
                 'item26c_red_lgth_assist':'other_Reduction Based on Length of Receipt of Assistance',
                 'item26c_other_non_sanc':'Other, Non-sanction'},axis='columns')



dff['category'] = ['Tea' if x==20
                  else('WorkPays' if x==84
                      else('Not Found')) for x in dff['category']]

# Funding Stream

dff['funding_stream'] = ['Funded, in whole or in part, with Federal TANF block grant funds. \
This includes programs funded with commingled State and Federal funds.' if x==1
                        else('Funded entirely from State-only funds (segregated State TANF \
                        expenditures) which are subject to most, but not all, TANF rules' if x==2
                            else('Not Found')) for x in dff['funding_stream']]

# Disposition

dff['disposition'] = ['Data collection completed' if x==1
                     else('Not subject to data collection/listed in error' if x==2
                         else('Not Found')) for x in dff['disposition']]

# New Applicant

dff['new_applicant'] = ['Yes, a newly-approved application' if x==1
                       else('Not newly-approved application' if x==2
                           else('Not Found')) for x in dff['new_applicant']]

# Type of Family for Work Participation

dff['family_type'] = ['Family included only in overall work participation rate \
(i.e., family includes one or more workeligible individuals but does not meet definition of a twoparent family' if x==1                                                
                      else('Two-Parent Family included in both the overall and\
                      two-parent work participation rates (i.e., family includes\
                      two work-eligible parents and meets the definition of a two-parent family)' if x==2                       
                           else('Family with no work-eligible individual' if x==3
                                else 'Not found')) for x in dff['family_type']]

# Receives Subsidized Housing

dff['subsid_housing'] = ['Public housing' if x==1
                        else('Rent subsidy' if x==2
                            else('No housing subsidy' if x==3
                                else('Not Found'))) for x in dff['subsid_housing']]

# Receives Medical Assistance

dff['medical_assist'] = [' Yes, enrolled in Medicaid and/or CHIP' if x==1
                        else('Does not receive Medical Assistance' if x==2
                            else('Not Found')) for x in dff['medical_assist']]

# Receives Food Stamps

dff['food_stamps'] = ['Yes, receives food stamp assistance' if x==1
                     else('Does not receive food stamp' if x==2
                         else('Not Found')) for x in dff['food_stamps']]

# Receives Subsidized Child Care

dff['subsid_child_care'] = ['Yes, receives child care funded entirely or in part with \
Federal funds (e.g., receives TANF, CCDF, SSBG, or other federally funded child care)' if x==1
                           else('Yes, receives child care funded entirely under a State, \
                           Tribal, and/or local program (i.e., no Federal funds used)' if x==2
                               else('No subsidized child care received' if x==3
                                   else('Not Found'))) for x in dff['subsid_child_care']]

# Work Requirements Sanction

dff['reason for amount of reduction_work requirements sanction'] = ['Yes' if x==1
                                                                   else('No' if x==2
                                                                   else('Not Found')) for x in dff['reason for amount of reduction_work requirements sanction']]

# Family Sanction for an Adult with No High School Diploma or Equivalent

dff['reason for amount of reduction_Family Sanction for an Adult with No High School Diploma or Equivalent'] = ['Yes' if x==1
                                                                   else('No' if x==2
                                                                   else('Not Found')) for x in dff['reason for amount of reduction_Family Sanction for an Adult with No High School Diploma or Equivalent']]

# Sanction for Teen Parent not Attending School

dff['reason for amount of reduction_Sanction for Teen Parent not Attending School'] = ['Yes' if x==1
                                                                   else('No' if x==2
                                                                   else('Not Found')) for x in dff['reason for amount of reduction_Sanction for Teen Parent not Attending School']]

# Non-Cooperation with Child Support

dff['reason for amount of reduction_Non-Cooperation with Child Support'] = ['Yes' if x==1
                                                                   else('No' if x==2
                                                                   else('Not Found')) for x in dff['reason for amount of reduction_Non-Cooperation with Child Support']]

# Failure to Comply with an Individual Responsibility Plan

dff['reason for amount of reduction_Failure to Comply with an Individual Responsibility Plan'] = ['Yes' if x==1
                                                                   else('No' if x==2
                                                                   else('Not Found')) for x in dff['reason for amount of reduction_Failure to Comply with an Individual Responsibility Plan']]

# Other Sanction

dff['reason for amount of reduction_Other Sanction'] = ['Yes' if x==1
                                                                   else('No' if x==2
                                                                   else('Not Found')) for x in dff['reason for amount of reduction_Other Sanction']]

# Family Cap

dff['other_Family Cap'] = ['Yes' if x==1
                           else('No' if x==2
                                else('Not Found')) for x in dff['other_Family Cap']]

# Reduction Based on Length of Receipt of Assistance

dff['other_Reduction Based on Length of Receipt of Assistance'] = ['Yes' if x==1
                                                                  else('No' if x==2
                                                                      else('Not Found')) for x in dff['other_Reduction Based on Length of Receipt of Assistance']]

# Other, Non-sanction

dff['Other, Non-sanction'] = ['Yes' if x==1
                           else('No' if x==2
                                else('Not Found')) for x in dff['Other, Non-sanction']]

# Is the TANF Family Exempt from the Federal Time-Limit Provisions

dff['exempt_time_limit'] = ['Family is not exempt from Federal time limit.' if x==1
                           else(' Yes, family is exempt from accrual of months and \
                           termination of assistance under the Federal five-year time\
                           limit for the reporting month because no headof-household \
                           or spouse of the head-of-household in the eligible family is receiving assistance.' if x==2
                               else('Yes, family is exempt from accrual of months under\
                               the Federal five-year time limit for the reporting month \
                               because assistance to family is funded entirely from \
                               State-only funds (See data element #8, Funding Stream, code 2.)' if x==3
                                   else('Yes, family is exempt from accrual of months \
                                   under the Federal five-year time limit for the reporting \
                                   month because the family is living in Indian country\
                                   or an Alaskan native village, where at least 50 percent \
                                   of the adults living in the Indian country or Alaskan native village are not employed.'if x==4
                                       else(' Code no longer in use' if x==5
                                           else('Yes, family is exempt from termination of assistance\
                                           under the Federal fiveyear time limit for the reporting month\
                                           because assistance to the family is funded entirely from \
                                           State-only funds (See data element #8, Funding Stream, code 2.)' if x==6
                                               else('Yes, family is exempt from termination of assistance \
                                               under the Federal fiveyear time limit for the reporting month \
                                               due to a hardship exemption, battery, or extreme cruelty \
                                               (i.e., victims of domestic violence who were not given a Federally \
                                               recognized domestic violence waiver).' if x==7
                                                   else(' Yes, family is exempt from termination of assistance under \
                                                   State policy for the reporting month based on a federally recognized \
                                                   good cause domestic violence waiver of time limits.' if x==8
                                                       else('Yes, family is exempt from termination of assistance under the \
                                                       Federal fiveyear time limit for the reporting month because the \
                                                       head-of-household or the spouse of the head-of-household is living \
                                                       in Indian country or an Alaskan native village, where at least 50 \
                                                       percent of whose adults are not employed.' if x==9
                                                           else('Code no longer in use' if x==10
                                                               else('Code no longer in use' if x==11
                                                                   else('Not Found'))))))))))) for x in dff['exempt_time_limit']]

# Is the TANF Family A New Child-Only Family?

dff['new_child_only_fam'] = ['Yes, a new child-only family' if x==1
                            else(' No, not a new child-only family'if x==2
                                else('Not Found')) for x in dff['new_child_only_fam']]


dff1 = dff.copy()
writer = pd.ExcelWriter(reporting_month + '-' + 't_worc_families' + '-' + timestamp + '.xlsx')
dff1.to_excel(writer, index = False)
writer.save()
```
